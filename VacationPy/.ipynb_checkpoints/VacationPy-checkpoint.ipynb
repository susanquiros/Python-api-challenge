### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
#defyning the locations 
weather_locations = clean_city[['Latitude', 'Longitude']]

weather_humidity= clean_city['Humidity %']
```

```python
fig_layout ={'width': '700px',
    'height': '700px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}


fig= gmaps.figure(layout= fig_layout)
# Create heat layer
heat_layer = gmaps.heatmap_layer(weather_locations, weights=weather_humidity, 
                                 dissipating=False, max_intensity=10,
                                 point_radius=1)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
hotel_df= clean_city.loc[(clean_city['Max Temperature']>=18) & (clean_city['Max Temperature']<=30) & (clean_city['Wind Speed (km/hr)']<=16) & (clean_city['Cloudiness %']== 0),:]
                                                                                                                                        
hotel_df.reset_index(drop=True)
#getting ride of the NaN- in my case I do not have those
#hot_weather.dropna(subset=['Country'], inplace = True)
#hot_weather.reset_index(drop=True)
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotels=[]

# Loop through dataframe to get nearest hotel
for city in range(len(hotel_df["City"])):
    lat = hotel_df.iloc[city]["Latitude"]
    lng = hotel_df.iloc[city]["Longitude"]
    city_coords = f"{lat},{lng}"
    params = {
        "location": city_coords, 
        "types": "lodging",
        "radius": 5000,
        "key": g_key
              }

    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"   
    hotel_response = requests.get(base_url, params=params).json()
 
    try:
        hotels.append(hotel_response["results"][0]["name"])
    except:
        hotels.append("Nearest hotel not found")

# Dataframe with nearest hotel
hotel_df["Hotel Name"] = hotels
hotel_df
```

```python
hotel_df.reset_index(drop=True)
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Latitude", "Longitude"]]
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations, info_box_content= hotel_info)
fig.add_layer(markers)
fig

# Display figure
```
